# Transformations

### 1. MAP

```python
x = sc.parallelize(["b", "a", "c"])
y = x.map(lambda a: (a,1))
print(x.collect())
print(y.collect())
```

### 2. Filter

```python
x= sc.parallelize([1,2,3])
y= x.filter(lambda x: x%2 == 1) #keep odd values
print(x.collect())
print(y.collect())
```

### 3. FlatMap

```python
x = sc.parallelize([1,2,3])
y = x.flatMap(lambda a: (a, a*100, 42))
z = x.map(lambda a: (a, a*100, 42))
print(x.collect())
print(y.collect())
print(z.collect())
```

### GroupBy

```python
x= sc.parallelize(['Python', 'Scala', 'Rlanguage', 'Spark'])
y = x.groupBy(lambda a: a[0])
#print(x.collect())
#print(y.collect())
print([(k,list(v)) for (k,v) in y.collect()])
#[(k, list(v)) for (k,v) in y.collect()]
```

# GroupByKey

```python
x= sc.parallelize([('B',5),('B',4),('A',3),('A',2),('A',1)])
y= x.groupByKey()
print(x.collect())
print(list((j[0], list(j[1])) for j in y.collect()))
```
